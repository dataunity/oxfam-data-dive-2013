```python
# Pivots the data so it's suitable for reporting in D3
# NOTE: Run the notebook 'Oxfam oil header conversion' before this one
import pandas as pd
from pandas import *

month_map = {
    'jan': '01',
    'feb': '02',
    'mar': '03',
    'apr': '04',
    'may': '05',
    'jun': '06',
    'jul': '07',
    'aug': '08',
    'sep': '09',
    'oct': '10',
    'nov': '11',
    'dec': '12'
}
```

```python
# Load data
dir_path = 'Data/Transformed/Production by month/output/'
filename = 'Production_of_Crude_Oil_including_Lease_Condensate_(Thousand_Barrels_Per_Day)_merged_header.csv'
output_filename = 'Production_of_Crude_Oil_including_Lease_Condensate_(Thousand_Barrels_Per_Day)_merged_header_pivoted.csv'
filepath = dir_path + filename
output_filepath = dir_path + output_filename
table = pd.read_csv(filepath, index_col='Country')
table_stacked = table.stack()
table_stacked = table_stacked.reset_index()
table_stacked.columns = ['Country', 'date', 'Value']
table_stacked[:10]
```

```python
# Group data by country
grouped_by_country = table_stacked.groupby([table_stacked['Country'], table_stacked['date']])
sum_by_country = grouped_by_country.sum()
sum_by_country[:10]
table_country = sum_by_country.reset_index()
table_country[:10]
```

```python
# Sort the dates
#table.Date.dtype.name

def f(x):
    year = x.split(' ')[1]
    #if int(year) > 50:
    #    year = '19' + year
    #else:
    #    year = '20' + year
    month = month_map[x.split(' ')[0].lower()]
    return '%s%s01' % (year, month)
#f = lambda x: '%s%s01' % (x.split('-')[1], month_map[x.split('-')[0]])
table_country['date'] = table_country['date'].apply(f)
table_country[:10]
```

```python
# Order by date
table_country = table_country.sort_index(by='date', ascending=False)
table_country[:10]
```

```python
date_by_country = table_country.pivot('date', 'Country', 'Value')
date_by_country.to_csv(output_filepath)
```
